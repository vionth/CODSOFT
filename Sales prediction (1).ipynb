##### Sales prediction using machine learning

```python
# import flowing packages
import pandas as pd
import numpy as np
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import seaborn as ad
import warnings
warnings.simplefilter("ignore")
```

```python
# insert the package of sales prediction
data=pd.read_csv("C:\\Users\\lenovo\\Downloads\\advertising.csv")
```

```python
#head()-calling the few datas in the dataset
data.head()
```

```python
#null values in the data set
data.isna().sum()
```

```python
# in the dataset variables use spliting into x adn y, x-independent, y-dependent
X = data.drop(columns=['Sales'])
Y = data['Sales']
```

```python
# TV Advirtisement
ad.histplot(x='TV',data=data)
plt.title("TV Advirtisement")
```

```python
# Newspaper Advirtisement
ad.histplot(x='Newspaper',data=data)
plt.title("Newspaper Advirtisement ")
```

```python
# Radio Advirtisement
ad.histplot(x='Radio',data=data)
plt.title("Radio Advirtisement")
```

```python
# Heat map
plt.figure(figsize=(10,6))
ad.heatmap(data.corr(),annot=True,cmap='twilight')
plt.title("Heat for data Representation")
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=42)
```

```python
print("Shape of X_train:",X_train.shape)
print("Shape of X_test:",X_test.shape)
print("Shape of Y_train:",Y_train.shape)
print("Shape of Y_test:",Y_test.shape)
```

```python
# creating multiple regression
model=LinearRegression()
# fitting the model
model.fit(X_train,Y_train)
# predict the model
pred=model.predict(X_test)
```

```python
# predicting the output by random inputs
print("Prediction 1:",model.predict([[220.3,24.2,55.9]]))
print("prediction 2:",model.predict([[17.2,35.9,59.3]]))
```

```python
# checking the accuracy of the model
from sklearn.metrics import r2_score
print("Accuracy_score:",r2_score(Y_test,pred))
```
